## 1. The YT tool: Turbulent cloud simulation (FLASH code)

In this section we will look into YT examples.

The package is here: https://yt-project.org

```python
# Import yt
import yt
import matplotlib.pyplot as plt
```

```python
ds = yt.load_sample("DensTurbMag")
```

```python
#ds.print_stats()
```

```python
ds.field_list
```

```python
#ds.derived_field_list
```

### Density map:

```python
# Save the plot with a specified figure size
p = yt.SlicePlot(ds, "y", ("gas", "density"))

p.set_figure_size(3)

p.set_cmap(("gas", "density"), "viridis")

p.show()
```

```python
# Save the plot with a specified figure size
p = yt.ProjectionPlot(ds, "y", ("gas", "density"))

p.set_figure_size(3)

p.set_cmap(("gas", "density"), "magma")

p.show()
```

### Temperature map:

```python
# Save the plot with a specified figure size
p = yt.SlicePlot(ds, "y", ("gas", "temperature"))

p.set_figure_size(3)

p.set_cmap(("gas", "temperature"), "magma")

p.show()
```

### Zoom-in:

```python
p.zoom(2.0)
p.annotate_velocity()
p.set_figure_size(3)
```

```python

p = yt.SlicePlot(ds, "x", ("gas", "density"), center="max")
p.set_figure_size(3)
p.annotate_contour(("gas", "density"))
p.zoom(2.5)
```

### Import H5PY, which is a library for handling HDF5 files.

```python
# Importing libraries

import h5py
import matplotlib.pyplot as plt
```

```python
# Empty list for subhalo positions
subhalo_coord = []

# I/O HDF5
with h5py.File("./groups_135.0.hdf5", 'r') as f:
    # Positions are typically in the 'Subhalo' group, 'SubhaloPos' dataset
    # Units are often in simulation box units (e.g., Mpc/h)
    subhalo_coord = f['Subhalo']['SubhaloPos'][:]
```

### Check data:

```python
# Check type
#print(type(subhalo_coord))

# The total number of subhalos per coordinate (x,y,z)
#print(subhalo_coord.shape)
```

### Plotting with Matplotlib (2D projection):

```python
# Create the map plot
plt.figure(figsize=(6,6))

plt.scatter(subhalo_coord[:, 0], subhalo_coord[:, 1], s=1, alpha=0.5, color='blue')

# Axes
plt.xlabel('X Position (Mpc/h)')
plt.ylabel('Y Position (Mpc/h)')

plt.grid(True, linestyle='--', alpha=0.4)

plt.show()
```
